```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
company = "Google"
company.lower()
# You can use this later to gather the data, by simply naming the files as the company name.
```

```python
df = pd.read_csv(f"{company}.csv")
```

```python
df = df.set_index(df[0:]['Year'])  #Removing the index 0,1,2,3 etc. So we can search easier later on.
```

```python
df = pd.DataFrame(df)
```

```python
df = df.replace('- -', np.nan)
```

```python
df = df.drop(['Year'], axis=1)
```

```python
df = df.drop(['FALSE', 'SEC Link'])
```

```python
df = df.fillna(0)
```

```python
goog = {
    {"Operating income": [df.loc['Operating Income']]},
    {"Net Receivables": [df.loc['Net Receivables']]},
    {"Inventory": [df.loc['Inventory']]},
    {"Other Current Assets": [df.loc['Other Current Assets']]},
    {"Accounts Payable": [df.loc['Accounts Payable']]},
    {"Deferred Revenue": [df.loc['Deferred Revenue']]},
    {"Other Current Liabilities": [df.loc['Other Current Liabilities']]},
    {"PP&E": [df.loc['PP&E']]},
    }
```

```python
gg1 = pd.DataFrame(data=goog)
gg1 = gg1.transpose()
gg1 = gg1.astype(float)/1000000
gg1
```

```python
gg1.loc['PP&E']
```

```python
nta = []
for i in range(len(gg1.loc['EBIT'])):
    nta.append(gg1.loc['Receivables'][i] + gg1.loc['Inventory'][i] + gg1.loc['OtherAssets'][i] + gg1.loc['PP&E'][i]
               - gg1.loc['AccountsPayable'][i] - gg1.loc['DeferredRevenue'][i] - gg1.loc['OtherLiabilities'][i])
    
nta = pd.Series(nta)
```

```python
ebit = gg1.loc['EBIT']
ppie = gg1.loc['PP&E']
x_ppie = [x for x in range(len(ppie))]
x_ebit = [x for x in range(len(ebit))]
x_nta = [x for x in range(len(nta))]
```

```python
ronta = []
for i in range(len(nta)):
    ronta.append(float((ebit[i]/nta[i])*100))
    
ronta = pd.Series(ronta)
ronta
```

```python
graph = plt.plot(ebit, label='Operating income')
graph = plt.plot(nta, label='Net Tangible Assets')
plt.xlabel('Years')
plt.ylabel("($ in millions)")
plt.title('Google')
plt.legend()
```

```python
fix, ax1 = plt.subplots()
ax1.bar(x_nta, nta, color = 'y', label = "Net tangible assets in millions")
plt.ylabel("Net tangible assets")
ax2 = ax1.twinx()
ax2.plot(ronta, color = 'g')
ax2.set_ylim([0, 200])
ax2.set_ylabel("% Return on net tangible assets")
plt.xlabel('Years')

plt.title(company)

#ax.plot(ebit, label = "Operating profit")
```
